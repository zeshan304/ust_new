# Create a Notebook in Amazon SageMaker
- Open the link link to console: https://console.aws.amazon.com/sagemaker/ 

- on the left side of the screen choose `notebook instances` from menu and then click on `Create notebook instance`  

- Supply notebook name and instance type and leave everything else unchanged and then press `Create notebook instance`

![](img/createnotebookinstance.png)



- your notebook status will be in pending
![](img/pendingstate.png)

- while the notebook is being created, you can go and create an S3 bucket.
- We use the bucket both for storing data as well ass saving the model.
- The bucketname should include the the word `sagemaker`. Amazon SageMaker by default has access to all buckets whosename include `sagemker`. This is only for training purposes and is not a good practice.
- Supply a bucket name and use the same region in which you created your notebook instance.
- leave everuthing else unchanged.

![](img/CreateBucket.png)

- Go back to sagemaker and check the status of your notebook instance.
- at somee point the status of your bucket would be `InService` and you can `open` your notebook.
![](img/notebookready.png)

- This will take you to your Jupyter Notebook.

- In the notebook instance choose `new` and then select `terminal`
- This will open a terminal window.
- in the terminal window glone this repository by executing: 

```cd ~/SageMaker
git clone https://github.com/cyrusmvahid/aim-418-labs.git```


- This will open your Jupyter environment.


- We are now ready to start [lab02](lab02-ndarray.ipynb).
